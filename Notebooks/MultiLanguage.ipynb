```python
import re
import spacy
import nltk
import json
from string import punctuation
from pykakasi import kakasi
```

```python
with open("array.json") as f:
    data = json.load(f)

japanese_regex = r'[\u4E00-\u9FFF\u30A0-\u30FF\u3040-\u309F]+'
english_regex = r'[a-zA-Z0-9\s]+'
# japanese_pieces = []
for token in data:
    english_part = ''.join(re.findall(english_regex, str(data)))
    japanese_part = ''.join(re.findall(japanese_regex, str(data)))

# print(re.findall(japanese_regex, str(data)))
# for token in data:
#     print(re.findall(japanese_regex, str(data)))
```

```python
def convert_kanji_to_hiragana(text):
    kakasi_inst = kakasi()
    kakasi_inst.setMode('J', 'H')  # Kanji to Hiragana mode
    converter = kakasi_inst.getConverter()
    hiragana_text = converter.do(text)
    return hiragana_text
```

```python
nlp_eng = spacy.load("en_core_web_sm")
nlp_jp = spacy.load("ja_core_news_sm")

# print(english_part)
# print(japanese_part)
for text in japanese_part:
    japanese_pieces.append(convert_kanji_to_hiragana(text))

print(english_part)
print(japanese_pieces)
doc_eng = nlp_eng(english_part)
doc_jp = nlp_jp(japanese_pieces)

# for token in doc_eng:
#     print(token)

tokens_eng=[tok.text for tok in doc_eng if tok.text not in punctuation]
print(tokens_eng)
tokens_jp=[tok.text for tok in doc_jp if tok.text not in punctuation]
print(tokens_jp)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_token_eng = []
for token in tokens_eng:
    stemmed_token_eng.append(stemmer.stem(token))
    print(token, "|", stemmer.stem(token))
```
