```python
import spacy
import nltk
import json
from string import punctuation
```

```python
nlp = spacy.load("en_core_web_sm")

with open("array.json") as f:
    data = json.load(f)

# json_stringify = json.dumps(data)
# print(json_stringify)

doc = nlp(str(data))
tokens = [tok.text for tok in doc if tok.text not in punctuation]

print (tokens)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
for token in tokens:
  print(token, "|", stemmer.stem(token))
```

```python
tokens_prepare = ' '.join(tokens)
doc = nlp(tokens_prepare)
for token in doc:
  print(token,"|",spacy.explain(token.tag_),"|", token.lemma_)
```
